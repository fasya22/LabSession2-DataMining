# One Hot Encode

```python
#Encode the itemset
itemset = set(items)
encoded_vals = []
for index, row in data.iterrows():
    rowset = set(row) 
    labels = {}
    uncommons = list(itemset - rowset)
    commons = list(itemset.intersection(rowset))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]
ohe_data = pd.DataFrame(encoded_vals)
```

```python
# Delete a column by column number
dataa = ohe_data.drop(columns=ohe_data.columns[133])
```

```python
databaru = dataa[2001 : 3001]
```

```python
databaru
```

# Applying Apriori

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
freq_items = apriori(databaru, min_support= 0.02, use_colnames=True)
```

* ohe_data adalah data yang akan dianalisa asosiasinya yang sudah di-encode
* min_support adalah minimum support yang akan ditentukan

```python
freq_items
```

```python
rules1 = association_rules(freq_items, metric="confidence", min_threshold=0.9)
```

* min_threshold menentukan minimum confidence yang akan dicari asosiasinya
* min_support ditentukan/diatur di freq_item yang ditentukan di atas

```python
rules1
```

```python
rules2 = association_rules(freq_items, metric="confidence", min_threshold=0.5)
```
